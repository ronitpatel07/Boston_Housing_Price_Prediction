# Data Preprocessing

```python
X=df_boston[['CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 'DIS', 'RAD', 'TAX',
       'PTRATIO', 'B', 'LSTAT']]
y=df_target['MedV']
```

```python
#Splitting Dataset into train and test data
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=101)
```

```python
#Linear regression model on train data
from sklearn.linear_model import LinearRegression
lm=LinearRegression()
lm.fit(X_train, y_train)
```

```python
# print the intercept
print(lm.intercept_)
```

```python
# print coefficient value of each attribute
coeff_df = pd.DataFrame(lm.coef_,X.columns,columns=['Coefficient'])
coeff_df
```

```python
#Making predictions on test data
predictions = lm.predict(X_test)
```

```python
#Comparing Actual data with predicted results
actual=pd.DataFrame(y_test.values, columns=['Actual'])
pred=pd.DataFrame(predictions, columns=['Predicted'])
disp=pd.concat([actual,pred], axis=1)
disp.head()
```

```python
plt.scatter(y_test,predictions)
plt.xlabel('Y Test')
plt.ylabel('Predicted Y')
plt.xlabel('Y Test')
plt.ylabel('Predicted Y')
```

```python
#Evaluating linear regression model with Mean Absolute Error, Mean Square Error and Root Mean Square Error
from sklearn import metrics
print('MAE:', metrics.mean_absolute_error(y_test, predictions))
print('MSE:', metrics.mean_squared_error(y_test, predictions))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, predictions)))
```
